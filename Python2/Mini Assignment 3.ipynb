1. Use for loop and range() function to create a list that includes even numbers from 50 to 100. Next create a list comprehension that does the same.

```
# Instantiate an empty list
range_list = []
# Skipping every other number to add every even number from 50 up to 100
for num in range(50, 101, 2):
  range_list.append(num)

# Same thing as above but a more condensed for loop inside of a list
comp_list = [num for num in range(50, 101, 2)]

# Print output
print(range_list)
print(comp_list)
```

2. Write a program to get two numbers from the user and creat a list that includes all the numbers between the two that are devisible by 3. Write your program with for loop once and another time with list comprehension.

```
# Gathering user input
first_num = int(input("Please enter a small whole number! "))
second_num = int(input("Please enter a large whole number! "))

# Variable instantiation
num_list = []

# For each number in the range from the first number to the second_number + 1
# to avoid any out of range errors
for num in range(first_num, second_num + 1, 3):
  num_list.append(num)
print(num_list)

# Same thing as above but more condensed in a list comprehension
list_of_nums = [num for num in range(first_num, second_num + 1, 3)]
print(list_of_nums)
```

3. Create a game robot named "Burnie". Burnie randomly selects a number between 1 and 100 and the user has 6 attempts to guess that number. At each user's guess, burnie tells the user if their guess is correct, or they should go higher or lower.

```
# Import relevant libraries
import random as rand

print("Hello, my name is Burnie. I have picked a secret number from 1 to 100 and it is your job to guess the number in 6 or less tries")
# Instantiate the relevant and necessary variables
computer_number = rand.randint(1, 100)
attempts = 6

# While loop so there is no fixed number of iterations the loop has to go through
while attempts > 0:
  user_guess = int(input("\nPlease make a guess to what Burnie's secret number "))
  # Win case
  if user_guess == computer_number:
    print("Congratulations, you guessed his number! ")
    break
  # Guess higher
  elif user_guess < computer_number:
    print("\nGuess higher next time! ")
  # Else guess lower
  else:
    print("\nGuess lower next time! ")

  attempts -= 1
  # Tell the user how many attempts they have left
  print(f"\nYou have {attempts} attempts left to guess Burnie's secret number ")

# Lose case
if attempts == 0:
  print(f"\nSorry, you've run out of attempts. Burnie's number was {computer_number}. Better luck next time! ")
```

4. Write a program to receive a movie genre from the user and creats **a list** of movie names from the dictionary below that include that specific genre. For instance, if the user inputs 'action', the program should display this list: ['Inception', 'Dark Knight', 'Gladiator']

 Write your program once with for loop and another time with list comprehension.

```
movies = {
    'Avatar':'Asabeneh',
    'Godfather':'crime|drama',
    'Inception':'action|adventure|sci-fi',
    'Dark Knight':'action|crime|drama',
    'Forrest Gump':'drama|romance',
    'Pianist':'biography|drama|music',
    'Back to the Future':'comedy|adventure|sci-fi',
    'Interstellar': 'comedy|adventure|sci-fi',
    'The Silence of the Lambs': 'crime|drama|thriller',
    'Gladiator':'action|adventure|drama'
}
```

```
# Gather user input and instantiate variables
user_input = input("Enter a movie genre: ")
movie_list = []

# For loop to iterate through the dictionary keys and values
for movie, genres in movies.items():
  # Spliting the genres from the | delimiter
  if user_input in genres.split('|'):
    movie_list.append(movie)
print(f"Movies found in that genre {movie_list}")

# Same for loop but inside of a list
list_movies = [movie for movie, genres in movies.items() if user_input in genres.split('|')]
print(f"Movies found in that genre {list_movies}")
```
