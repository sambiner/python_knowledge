## Bank Account Problem
* Create a Python class called BankAccount which represents a bank account, having as attributes: accountNumber (numeric type), name (name of the account owner as string type), balance.
* Create a constructor method with parameters: accountNumber, name, balance.
* Create a deposit() method which manages the deposit actions. It receives the deposit amount as a parameter and updates the balance attribute.
* Create a withdrawal() method  which manages withdrawals actions. It should check the withdrawal to be less than the balance. If so, it should update the balance with the new amount.
* Create an bankfees() method to apply the bank fees with a percentage of 5% of the balance account. Every time this method is called, it should update the account balance to reflect the bank fee charge.
* Create a display() method to display account details.
* Create an object of the BankAccount class, and apply all the above methods to make sure they work correctly.

```
class BankAccount:
  def __init__(self, accountNumber: int, name: str, balance: int):
    self.accountNumber = accountNumber
    self.name = name
    self.balance = balance

  # Alters and returns the balance of the user's bankaccount depending on the amount inputted
  def deposit(self, amount: int):
    self.balance += amount
    return self.balance

  # Can only withdraw positive integers
  # When money is withdrawn, update and return the user's account balance
  def withdrawl(self, amount: int):
    if amount < 0:
      print("Withdrawl amount cannot be negative")
      return
    else:
      self.balance -= amount
      return self.balance

  # Update and return user's account balance based on 5% bank fees
  def bank_fees(self):
    self.balance = self.balance - (self.balance * 0.05)
    return self.balance

  # Simple display method that prints out all relevant account information
  def display(self):
    print(f"Account holder {self.name} has account {self.accountNumber} with a total balance of {self.balance}")

person = BankAccount(102038564853, "Arvin Mesgari", 100000) # Nice balance Professor!!
print(person.deposit(2500))
print(person.withdrawl(2050))
print(person.bank_fees())
person.display()
```

## Class List Problem
* Create a Python class called ClassList which represents a list of students in a class, having as attributes course name (string type), course number (string type), section number (number type), student names (list type) and student grades (list type).
* Create a constructor method with parameters: course_name, course_number, section_num, student_names, and student_grades
* Create a student_add() method which receives the name and grade of a new student and adds it to the student list.
* Create a grade_summary() method which displays the class grade average, minimum, and maximum.
* Create a top_student() method which displays the name of the student with the highest grade.
* Create a grade_update() method that receives the name of the student and their new grade, and updates their grade.


```
import math
class ClassList:
  def __init__(self, course_name: str, course_number: str, section_num: int, student_names: str, student_grades: int):
    self.course_name = course_name
    self.course_number = course_number
    self.section_num = section_num
    self.student_names = student_names
    self.student_grades = student_grades
    # Prints the initial class list of students and grades inputted
    print(f"Class consisting of students {self.student_names} with grades of {self.student_grades}")

  # Adds a students and their grade to the class list
  def student_add(self, name: str, grade: int):
    self.student_names.append(name)
    self.student_grades.append(grade)
    print("New student added!")

  # Returns the average, minimum, and maximum grades in the class
  def grade_summary(self):
    summary = [sum(self.student_grades)/len(self.student_grades), min(self.student_grades), max(self.student_grades)]
    print(f"Class average: {summary[0]}; Lowest in the class: {summary[1]}; Highest in the class: {summary[2]}")

  # Gets the index of the highest grade and prints out the student with that highest grade
  def top_student(self):
    best_grade_ind = self.student_grades.index(max(self.student_grades))
    print(f"The student with the highest grade is {self.student_names[best_grade_ind]}")

  # Find the index of the student name and then update that specific student's grade with the inputted value
  def grade_update(self, name: str, grade_update: int):
    name_index = self.student_names.index(name)
    self.student_grades[name_index] = grade_update
    print(f"{name}'s new grade is {grade_update}")

  # Simple display class to show new student names and updated grades
  def display_class(self):
    print(f"Class consisting of students {self.student_names} with grades of {self.student_grades}")

class_a = ClassList("Business Programming", "ISBA 3730", 1, ["John Doe", "Jane Doe", "Issac Newton", "Jack Linter"], [93,90,72,94])
class_a.student_add("Lightning McQueen", 85)
class_a.grade_summary()
class_a.top_student()
class_a.grade_update("Issac Newton", 97)
class_a.display_class()
```
