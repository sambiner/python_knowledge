17. Create a function that receives the name of an auto maker and returns a slice of the dataframe with the cars from that auto maker. For instance, it may receive 'ford' and return a slice with all the cars produced by 'ford' meaning they have 'ford' in their name. Call the function for 'ford' to see if it works properly.
> **Hint:** You may create a list comprehension producing a list of all the index labels that include the auto-maker's name inside them. Then you feed this list to .loc[ ] to slice for those index labels.

```
def filter_by_name(name):
  car_indices = [index for index, car_name in enumerate(mpg_df["name"]) if name.lower() in car_name.lower()]
  return mpg_df.loc[car_indices]

mpg_df.reset_index(drop=True, inplace=True)
filter_by_ford = filter_by_name("ford")
filter_by_ford
```

18. Create a function that receives a list of automakers and returns a dataframe with average mpg, horsepower, and cylinders for the cars produced by each auto maker in that list. For instance, the function may receive the list of ['ford', 'chevy', 'dodge'] and then return a dataframe with three rows (for the three auto makers) and three columns of mpg, horsepower, and cylinders to display the averages for each automaker.
> **Hint:** You can call the function created in previous question inside the definition of this function.



```
def car_averages(names):
  average_df = pd.DataFrame(columns=["name","average_mpg","average_horsepower","average_cylinders"])

  for name in names:
    filtered_names = filter_by_name(name)

    avg_mpg = filtered_names["mpg"].median()
    avg_hp = filtered_names["horsepower"].median()
    avg_cyl = filtered_names["cylinders"].median()

    average_df = average_df.append({
        "name": name,
        "average_mpg": avg_mpg,
        "average_horsepower": avg_hp,
        "average_cylinders": avg_cyl,
    }, ignore_index=True)

  return average_df

check_car_averages = car_averages(["ford", "chevy", "dodge"])
check_car_averages
```
