# Create Playlist

```python
# my personal Spotify user ID
user_id = 'sambmang'
playlist_endpoint_url = f'https://api.spotify.com/v1/users/{user_id}/playlists'
```

```python
request_body = json.dumps({
    'name':f'{source_artist_name} Mixtape',
    'description':f'Python generated playlist using artists related to {source_artist_name}',
    'public':False
})

# Post request
playlists_api_request = requests.post(playlist_endpoint_url, headers=headers, data=request_body)
```

```python
playlists_api_request
```

```python
playlists_api_response = json.loads(playlists_api_request.text)
playlists_api_response
```

```python
playlist_id = playlists_api_response['id']
```

```python
spotify_url = playlists_api_response['external_urls']['spotify']
spotify_url
```

# Get Artist Top Tracks & Insert into Playlist

```python
selected_track_uris = []

for artist in selected_related_artists:
    print(artist, selected_related_artists[artist])
    
    artist_id = selected_related_artists[artist]
    artist_top_tracks_endpoint_url = f'https://api.spotify.com/v1/artists/{artist_id}/top-tracks'
    
    # Same as the market on the Spotify Web API Console
    params = {'country':'US'}
    artist_top_tracks_api_request = requests.get(artist_top_tracks_endpoint_url, headers=headers, params=params)
    
    artist_top_tracks_api_response = json.loads(artist_top_tracks_api_request.text)
    for song in artist_top_tracks_api_response['tracks'][0:5]:
        print(artist, song['name'], song['uri'])
        selected_track_uris.append(song['uri'])
    print('-'*125)
```

```python
artist_top_tracks_api_response
```

```python
selected_track_uris
```

```python
playlist_tracks_endpoint_url = f'https://api.spotify.com/v1/playlists/{playlist_id}/tracks'

request_body = json.dumps({
    'uris':selected_track_uris
})

playlist_tracks_api_request = requests.post(playlist_tracks_endpoint_url, headers=headers, data=request_body)
```
